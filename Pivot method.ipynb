```python
import pandas as pd
```

```python
df = pd.read_csv('salesmen.csv',parse_dates=['Date'])
```

```python
df.columns = df.columns.str.lower()
```

```python
df
```

```python
df['salesman'].value_counts()
```

```python
df['salesman'] = df['salesman'].astype('category')
```

```python
df.info()
```

```python
df.pivot(index='date',columns='salesman',values='revenue')
```

# pivot tables

```python
foods = pd.read_csv('foods.csv')
```

```python
foods
```

```python
foods.columns = foods.columns.str.lower().str.replace(" ","_")
```

```python
foods.pivot_table(values='spend',index=['gender','item'],columns=['city'],aggfunc='sum')
```

```python
pd.pivot_table(foods,values='spend',index='gender',columns='city')
```
