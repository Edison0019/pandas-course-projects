```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fortune1000.csv',index_col='Rank')
```

```python
df
```

```python
sector = df.groupby(by='Sector')
```

```python
df['Sector'].nunique()
```

```python
sector.size().sort_values(ascending=False)
```

```python
sector.first()
```

```python
sector.last()
```

```python
#returning the values and the indexes of each group as a dict
for i in sector.groups:
    print(i)
```

# get_group() method

```python
sector.get_group('Energy')
```

```python
sector.get_group('Apparel')
```

```python
sector.max()
sector.min()
```

```python
sector.sum()
sector.mean()
```

# Grouping by multiple columns

```python
sector_industry = df.groupby(by=['Sector','Industry'])
```

```python
sector_industry.size()
```

```python
sector_industry.mean()
```

# the agg() method

```python
sector.agg({"Revenue":"sum",'Profits':"mean"})
```

```python
sector.agg(['sum','mean','size'])
```

```python
sector.agg({'Profits':['mean','max']})
```

# iterating on a group

```python
max_sector = pd.DataFrame(columns=df.columns)
```

```python
max_sector
```

```python
for sectors,data in sector:
    v = data.nlargest(1,'Profits')
    max_sector = max_sector.append(v)
```
