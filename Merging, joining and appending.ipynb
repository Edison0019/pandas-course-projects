```python
import pandas as pd
```

```python
week1 =pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 =pd.read_csv('Restaurant - Week 2 Sales.csv')
customer = pd.read_csv('Restaurant - Customers.csv')
food = pd.read_csv('Restaurant - Foods.csv')
```

# contact method

```python
pd.concat([week1,week2],ignore_index=True)
```

```python
#how to add keys to the different dataframes
weeks = pd.concat(objs=[week1,week2],keys=['week1','week2'])
```

```python
weeks
```

```python
weeks.loc[('week1',2),'Customer ID']
```

# append method

```python
week1.append(week2,ignore_index=True)
```

# Inner joins

```python
week1.merge(week2,how='inner',on='Customer ID',suffixes=[' - 1',' - 2'])
```

# Merging with multiple keys

```python
week1.merge(week2,how='inner',on=['Customer ID','Food ID'])
```

```python
week1.merge(week2,how='inner',on=['Customer ID','Food ID'])
```

# outer joins

```python
merged = week1.merge(week2,how='outer',on='Customer ID',suffixes=[' - w1', ' - w2'],indicator=True)
```

```python
merged
```

```python
merged['_merge'].value_counts()
```

```python
merged[~merged['_merge'].isin(['both'])]
```

# right and left join 

```python
week1.merge(week2,on='Customer ID',how='right')[['Customer ID','Food ID_y']]
```

```python
food
```

```python
week1.merge(food,on='Food ID',how='left',sort=True)
```

# the left_on or right_on

```python
week2
```

```python
 customer
```

```python
week2.merge(customer,how='left',left_on='Customer ID', right_on='ID').drop(labels=['ID'],axis=1)
```

# merging by index

```python
#we need to specify if we are going to use 
#either the right or left index for making the relationship in both tables
week1.merge(week2,how='inner',left_index=True,right_index=True)
```
