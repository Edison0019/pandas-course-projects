```python
tips = pd.DataFrame(data=np.random.randint(1,25,(4,6)),index=['fri','sat','sun','mon'],columns=[1,2,3,4,5,6])
```

```python
tips
```

```python
tips.transpose().plot(kind='barh',alpha=0.5)
```

```python
tips = tips.div(tips.sum(axis=1).astype('float'),axis=0)
```

```python
tips.plot(kind='bar',stacked=True,title='Test',legend=True)
```

```python
comp1 = np.random.normal(0, 1, size=200) # N(0, 1)
comp2 = np.random.normal(10, 2, size=200) # N(10, 4)
values = pd.Series(np.concatenate([comp1, comp2]))
values.hist(bins=100, alpha=0.3, color='k')
values.plot(kind='kde', style='k--',secondary_y=True)
```
